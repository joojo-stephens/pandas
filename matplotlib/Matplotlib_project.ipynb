# Loading the necessary libraries 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
```

# Get rid of that text

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

# Adding a tittle

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)

plt.title("My fitst graph")

plt.show()
```

# Adding x label and y label

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)

plt.title("My fitst graph")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.show()
```

# Changingt fonts

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)

plt.title("My first graph", fontdict = {"fontname": "Times New Roman", "fontsize": 20})
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.show()
```

# Removing the decimals

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y)

plt.title("My first graph", fontdict = {"fontname": "Times New Roman", "fontsize": 20})
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.xticks([0,1,2,3, 4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```
